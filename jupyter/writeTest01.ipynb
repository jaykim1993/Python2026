```python
## score_data.csv 파일의 데이터 중 국어점수가 85점 이상인 데이터만 추출하여 출력하세요.
## DictReader => csv 파일을 읽기 좋게 변환하는데 첫 줄을 제목(=필드명)으로 인식하도록 변경해준다.
```

```python
import csv
# 1. 파일 읽기
with open('greenData/score_data.csv','r',encoding='utf-8-sig') as f:
    # 2. 첫줄 필드명 인식시키기
    reader = csv.DictReader(f)
    print('국어 점수가 85점 이상인 데이터: ')
    
    # 3. 한줄씩 확인
    for line in reader:
        # 파일에서 읽어온 모든 숫자는 문자열이므로 반드시 int()로 형변환 해야함.
        if int(line['국어']) >= 85:
            print(line)
```
