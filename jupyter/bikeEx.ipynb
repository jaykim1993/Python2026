```python
import pandas as pd

# 출력 옵션 설정
pd.set_option('display.max_columns',None) # 모든 컬럼 출력
pd.set_option('display.width',1000) # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width',True) # 한글 정렬 맞추기

# 공공데이터포털에서 다운로드 받은 csv 파일 읽어오는 방법
df = pd.read_csv('greenData/bike02.csv', encoding='cp949')
print(df)
```

```python
print("상위 5개 데이터")
print(df.head(5))
```

```python
print('대여건수 1000 이상')
print(df[df['대여건수']>170000])
```

```python
df.sort_values(by='대여건수', ascending=False)
```

```python
total = df['대여건수'].sum()
print('전체 대여건수 합계: ', total)
avg = df['대여건수'].mean()
print(f'평균 대여건수: {avg:.1f}')
```

```python
# 내림차순 정렬 후, 첫 번째(0번 index) 행의 '대여일자' 출력
top_date = df.sort_values(by='대여건수', ascending=False).iloc[0]['대여일자']
print('가장 대여 많은 날짜: ', top_date)
```
