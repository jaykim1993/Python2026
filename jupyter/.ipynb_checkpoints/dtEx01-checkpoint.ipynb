```python
import pandas as pd

data = {
    '대여소': ['강남', '강남', '홍대', '홍대', '잠실', '잠실'],
    '이용자수': [120, 150, 80, 90, 200, 180],
    '대여시간': [30, 45, 20, 25, 60, 55],
    '이용일자': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02', '2024-01-01', '2024-01-02']
}

df = pd.DataFrame(data)


# 상위 3개 데이터 출력
# df.head(3) 
# df.info()
# 조건 필터
# df[df['이용자수'] >= 100]
# df.sort_values(by='대여시간', ascending=True)
# asc의 경우 생략해도 디폴트
# 내림차순 정렬 (by="", ascending=False)
```

```python
df['총이용량']=df['이용자수']*df['대여시간']
df
```

```python
# Data자료를 이용해 대여소별 총이용자수를 구한 후 
# 이용자수 많은 순으로 정렬하여 출력하시오
df.groupby('대여소')['이용자수'].sum().sort_values(ascending=False)
```

```python
# Data자료를 이용해 이용자수가 100 이상인
# 데이터만 추출하여 대여소별 평균대여시간 구하여 출력하시오
result = df[df['이용자수'] >= 100]
result.groupby('대여소')['대여시간'].mean()
```

```python
# Data자료를 이용해 대여소별 "총이용량" 합계 구하기
print("대여소별 총이용량 합계")
print(df.groupby('대여소')['총이용량'].sum())
```

```python
# 가장 바쁜 날짜 찾기.
busyDate = df.groupby('이용일자')['이용자수'].sum().sort_values(ascending=False)
print(busyDate)
print("가장 바쁜 날짜", busyDate.index[0])
```

```python
# 이용자수 평균이 가장 높은 대여소 찾기.
line = df.groupby('대여소')['이용자수'].mean().sort_values(ascending=False)
print(line)
print(f"평균 이용자수 가장 높은 대여소: {line.index[0]}")
```
