```python
# 반드시 pandas는 import하여 사용
import pandas as pd
import matplotlib.pyplot as plt

# 출력 옵션 설정
pd.set_option('display.max_columns', None)  # 모든 컬럼 출력
pd.set_option('display.width', 1000)  # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width', True)  # 한글 정렬 맞추기
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.tight_layout()
# 파일 불러오기
df = pd.read_csv('greenData/seoul.csv', encoding='cp949')
# NaN이 존재하는 행 모두 제거
df = df.dropna()
df.isnull().sum()

# 연도별 평균 미세먼지 농도 계산
result = df.groupby('측정년도')['미세먼지(㎍/㎥)'].mean()

plt.figure(figsize=(10, 6))

plt.bar(result.index, result.values)
# plt.xlim(2011.5, 2026.5)
plt.title('측정년도별 미세먼지 평균')
plt.xlabel('측정년도')
plt.ylabel('미세먼지 평균값')
plt.grid(linestyle=':')
plt.show()
```

```python
# 연도별 평균 미세먼지 농도 계산
result = df.groupby('측정년도')['초미세먼지(㎍/㎥)'].mean()
# result
plt.figure(figsize=(10, 6))
result.plot(kind='line', marker='o')
plt.title('측정년도별 초미세먼지 변화 (Seaborn)')
plt.xlabel('측정년도')
plt.ylabel('초미세먼지 평균값')
plt.grid(linestyle=':')
plt.legend()
plt.show()
```
