```python
# 병합(Merge): 최소 2개 이상의 데이터프레임을 합치는 작업
# 예시 데이터
import pandas as pd

# 예시 데이터
df1 = pd.DataFrame({
    "지역": ["서울", "부산", "대구"],
    # "연도": [2024,2025,2024],
    "총인구": [1000000, 300000, 200000]
})

df2 = pd.DataFrame({
    "지역": ["서울", "부산", "인천"],
    # "연도": [2024,2025,2026],
    "병원수": [150, 80, 60]
})

# 확인
df1, df2
# 데이터프레임이름.columns - 데이터프레임의 필드명을 확인하는 함수
print(df1.columns)
print(df2.columns)
```

```python
# pd.merge(Left_DataFameName, Right_DataFameName, how='타입', on='기준컬럼(PK,FK)')
# how 방식: inner / left / right / outer(합집합)
# inner
# merged = pd.merge(df1, df2, how='inner', on='지역')
# outer 합집합
# merged = pd.merge(df1, df2, how='outer', on='지역')
# left 왼쪽 존재하면 출력
# merged = pd.merge(df1, df2, how='left', on='지역')

# 컬럼명이 다를 경우
# merged = pd.merge(df1, df2, left_on='지역', right_on='지역명', how='inner')

# 여러 컬럼 기준으로 병합
# merged = pd.merge(df1, df2, how='inner', on=['지역', '연도'])
# print(merged)
```

```python
# 결측치: 데이터 병합과정에서 데이터가 비어있는 상태
# NaN으로 출력됨, 이후 연산에 걸림돌
# 결측치 처리 방법
# 1. fillna(): 특정값으로 채우기 - 0 또는 평균값
merged_df = pd.merge(df1, df2, how='outer', on='지역')
# print(merged_df.isna().sum())
# merged_df['총인구'] = merged_df['총인구'].fillna(0)
merged_df['총인구'] = merged_df['총인구'].fillna(merged_df['총인구'].mean())
# merged_df['병원수'] = merged_df['병원수'].fillna(0)
merged_df['병원수'] = merged_df['병원수'].fillna(merged_df['병원수'].mean())
merged_df['비율'] = merged_df['총인구']/merged_df['병원수']
# 2. dropna(): 결측치가 있는 행 제거
# 비추..
# droped_df = merged_df.dropna()
merged_df.style.format({'총인구': '{:,.0f}', '병원수': '{:,.0f}', '비율': '{:,.0f}'})

# print(droped_df)
```
