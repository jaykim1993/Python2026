```python

# 반드시 pandas는 import하여 사용
import pandas as pd
import matplotlib.pyplot as plt
# matplotlib기반에서 차트 예쁘게 꾸미는 도구
import seaborn as sns

# 한글 깨짐 방지 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

data = pd.DataFrame({
    "수학": [80, 90, 70],
    "영어": [86, 60, 75],
    "과학": [90, 88, 95],
}, index=['김학생', '이학생', '박학생'])

# 히트맵: 데이터 값을 색으로 표현하는 그래프
# 숫자가 클수록 진한색

sns.heatmap(data)
plt.title("학생 성적 히트맵")
plt.show()
```
