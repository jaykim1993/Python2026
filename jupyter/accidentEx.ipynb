```python
import pandas as pd

# 출력 옵션 설정
pd.set_option('display.max_columns',None) # 모든 컬럼 출력
pd.set_option('display.width',1000) # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width',True) # 한글 정렬 맞추기

# 공공데이터포털에서 다운로드 받은 csv 파일 읽어오기
df = pd.read_csv('greenData/accident.csv', encoding='cp949')
df
```

```python
# 날짜 추출하는 함수 종류
# CSV에 입력된 날짜는 문자열 데이터 타입이다.
# 날짜 데이터 형식으로 변환 : datetime()
df['사고일자'] = pd.to_datetime(df['사고일자'])
```

```python
# dataTime에서 필요 값 추출
# dt.year / dt.month / dt.day / dt.hour / dt.minute / dt.second
# dt.dayofweek - 요일을 숫자로 반환(0:월 ~ 6:일)
# dt.day_name() - 'Monday'..
# dt.quarter - 분기 
df['사고월'] = df['사고일자'].dt.month
```

```python
# .count(): 데이터가 비어있지 않은 값의 개수를 센다.
# .size(): 그룹별 전체 행의 개수를 센다.
monthly_accident = df.groupby('사고월').size()
print(monthly_accident)

# .idxmax(): 가장 큰 값이 들어있는 index
# most_accident_month = monthly_accident.sort_values(ascending=False).index[0]
maxMonth = monthly_accident.idxmax()
# print(f'가장 사고가 많은 월: {most_accident_month}월' )
print(f'가장 사고가 많은 월: {maxMonth}월' )
```

```python
print('--- 사상자수 기준 상위 5개 위험 노선 ---')
df['사상자수'] = df['사망'] + df['부상']
monthly_insured = df.groupby('노선명')['사상자수'].sum()
monthly_insured_sorted = monthly_insured.sort_values(ascending=False).head(5)
print(monthly_insured_sorted)
```
